# **Algoritmo Divide y Vencerás (Divide and Conquer Algorithm)**

# Minimo y Maximo Valor de una Lista (Arreglo)

# Usando Enfoque Clásico
## Por ejemplo Fuerza Bruta (Brute-Force)

```python
def max_arr(arr):
  max = arr[0]
  for i in range(1, len(arr)):
    if(arr[i] > max):
      max = arr[i]
  return max
```

```python
def min_arr(arr):
  min = arr[0]
  for i in range(1, len(arr)):
    if(arr[i] < min):
      min = arr[i]
  return min
```

```python
import random

def main():
  arr = list(range(30)) #Genera elementos ordenados [0,1,2,...,29] 
  #print(arr)          
  random.shuffle(arr)   #Barajar la secuencia
  print("Arreglo: ", arr)

  print("Minimo: ", min_arr(arr))
  print("Maximo: ", max_arr(arr))

main()
```

# Usando Divide y Vencerás

Solucion 01

```python
def obtenerMax(arr, i, j):
  if i == j:
    return arr[i]

  med = (i+j) // 2
  maxi = obtenerMax(arr, i, med)
  maxd = obtenerMax(arr, med + 1, j)
  return maxi if maxi > maxd else maxd


def obtenerMin(arr, i, j):
  if i == j:
    return arr[i]

  med = (i+j) // 2
  mini = obtenerMin(arr, i, med)
  mind = obtenerMin(arr, med + 1, j)
  return mini if mini < mind else mind 
```

```python
import random

def main():
  arr = [random.randint(1,200) for x in range(25)]
  print(arr)

  print("El minimo valor en el arreglo es: ", obtenerMin(arr, 0, len(arr)-1))
  print("El maximo valor en el arreglo es: ", obtenerMax(arr, 0, len(arr)-1))

main()
```

Solucion 02

```python
def divideAndConquer_Max(arr, ind, len):
    maximum = -1;

    if (ind >= len - 2):  #Si llegamos el penultimo indice
        if (arr[ind] > arr[ind + 1]):
            return arr[ind];
        else:
            return arr[ind + 1];

    maximum = divideAndConquer_Max(arr, ind + 1, len);

    if (arr[ind] > maximum):
        return arr[ind];
    else:
        return maximum;


def divideAndConquer_Min(arr, ind, len):
    minimum = 0;
    if (ind >= len - 2):  #Si llegamos el penultimo indice
        if (arr[ind] < arr[ind + 1]):
            return arr[ind];
        else:
            return arr[ind + 1];

    minimum = divideAndConquer_Min(arr, ind + 1, len);

    if (arr[ind] < minimum):
        return arr[ind];
    else:
        return minimum;
```

```python
def main():
  minimum, maximum = 0, -1;

  # Inicializamos el arreglo
  arr = [6, 4, 8, 90, 12, 56, 7, 1, 63];

  maximum = divideAndConquer_Max(arr, 0, 9);
  minimum = divideAndConquer_Min(arr, 0, 9);

  print("El minimo numero en el arreglo es: ", minimum);
  print("El maximo numero en el arreglo es: ", maximum);

main()
```
