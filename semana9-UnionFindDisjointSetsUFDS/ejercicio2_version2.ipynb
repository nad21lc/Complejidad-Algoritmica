```python
import graphviz as gv

#generar el grafo

def drawUF(myList):
    dot = gv.Digraph(comment="UFDS")
    lon = len(myList)
    
    for i in range(lon):
        #agreglo los nodos al grafo (valor, etiqueta que se va a mostrar en el grafo)
        dot.node(str(i), str(i)) #tienen que ser cadenas
        
    for j in range(lon):
        #en la arista pongo (valor del indice, valor del padre)
        if myList[j] >= 0: #Si ponderacion >= 0
            dot.edge(str(j), str(myList[j]))
    
    dot.graph_attr["rankdir"] = "BT" #que se muestre como arbol binario
    return dot


```

```python
# RetornarÃ¡ la ponderacion del nodo
def quickUnionPonderado_find(lista, buscado):
    i = buscado
    while lista[i] >= 0:
        i = lista[i]
        #print(i)
    return i
```

```python
#balancea segun el peso
def quickUnionPonderado_union(lista, a, b): # lista de ponderaciones
    pesoA = quickUnionPonderado_find(lista, a)
    pesoB = quickUnionPonderado_find(lista, b) 
    print("  pa =",pesoA, ",    pb =",pesoB)
    if pesoA == pesoB: 
        return
    print("lista[pa]=",lista[pesoA], ", lista[pb]=",lista[pesoB])
    if lista[pesoA] < lista[pesoB]:
        lista[pesoA] += lista[pesoB]
        lista[pesoB] = pesoA
        print ("Ponderacion de p < ponderacion de q, entonces, nuevas ponderaciones para la raiz ", pesoA,  "p=", lista[pesoA], "q= ", pesoA)
    elif lista[pesoB] < lista[pesoA]:
        lista[pesoB] += lista[pesoA]
        lista[pesoA] = pesoB
        print ("Ponderacion de p > ponderacion de q, entonces, nuevas ponderaciones para la raiz ", pesoB,  "p=", lista[pesoB], "q= ", pesoB)
    else:
        lista[pesoA] += lista[pesoB]
        lista[pesoB] = pesoA
        print ("Igual ponderacion de p y q, entonces, nuevas ponderaciones", "p=", lista[pesoA], "q=", lista[pesoB])
```

```python
rank = [-1 for i in range(10)]
print(rank)

dot = drawUF(rank)
dot
```
