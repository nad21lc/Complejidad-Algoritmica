```python
import graphviz as gv

#generar el grafo

def drawUF(myList):
    dot = gv.Digraph(comment="UFDS")
    lon = len(myList)
    
    for i in range(lon):
        #agreglo los nodos al grafo (valor, etiqueta que se va a mostrar en el grafo)
        dot.node(str(i), str(i)) #tienen que ser cadenas
        
    for j in range(lon):
        #en la arista pongo (valor del indice, valor del padre)
        if j != myList[j]:
            dot.edge(str(j),str(myList[j]))
    
    dot.graph_attr["rankdir"] = "BT" #que se muestre como arbol binario
    return dot
```

```python
def quickUnionPonderado_find(lista, a):
    while lista[a] != a:
        lista[a] = lista[lista[a]]  # Compresión de caminos
        a = lista[a]
    return a
```

```python
#[Union] Une los grupos usando a uno de los padres como padre común
def quickUnionPonderado_union(lista, a, b):
    raizA = quickUnionPonderado_find(lista, a)
    raizB = quickUnionPonderado_find(lista, b)
    
    if raizA == raizB:
        return
    
    if lista[raizA] < lista[raizB]:
        lista[raizB] = raizA
    else:
        lista[raizA] = raizB
```

```python
rank = []

for i in range(10):
    rank.append(i)

dot = drawUF(rank)
dot
```
