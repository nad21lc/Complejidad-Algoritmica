```python
import graphviz as gv

#generar el grafo

def drawUF(myList):
    dot = gv.Digraph(comment="UFDS")
    lon = len(myList)
    
    for i in range(lon):
        #agreglo los nodos al grafo (valor, etiqueta que se va a mostrar en el grafo)
        dot.node(str(i), str(i)) #tienen que ser cadenas
        
    for j in range(lon):
        #en la arista pongo (valor del indice, valor del padre)
        if j != myList[j]:
            dot.edge(str(j),str(myList[j]))
    
    dot.graph_attr["rankdir"] = "BT" #que se muestre como arbol binario
    return dot
```

```python
arr = [0,0,9,9,9,6,6,7,8,9]
drawUF(arr)
```

```python
#[Find] Busca si los padres son diferentes para unirlo
def quickfind_find(lista, buscado):
    return lista[buscado]
```

```python
lista = [0,0,9,9,9,6,6,7,8,9]
padre = quickfind_find(lista, 3)
print(padre)
padre = quickfind_find(lista, 5)
print(padre)
```

```python
#[Union] Une los grupos usando a uno de los padres como padre com√∫n.
def quickFind_union(lista,a,b):
    padreA = quickfind_find(lista, a)
    padreB = quickfind_find(lista, b)
    for i in range(len(lista)):
        if lista[i] == padreA:
            lista[i] = padreB
```

```python
arrTemp = [i for i in range(9)]
dot = drawUF(arrTemp)
dot
```

```python

a = quickFind_union(arrTemp, 3, 4)
dot = drawUF(arrTemp)
dot
```
